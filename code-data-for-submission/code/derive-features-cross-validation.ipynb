```python
# Required libraries
import numpy as np
import pandas as pd
from collections import Counter
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFECV
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_selection import mutual_info_classif

# ------------------------ Resampling Function ------------------------ #
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import (ClusterCentroids, EditedNearestNeighbours, 
                                     RepeatedEditedNearestNeighbours, InstanceHardnessThreshold, RandomUnderSampler)

def apply_resampling(X, y, method='smote'):
    if method == 'smote':
        sampler = SMOTE(random_state=42)
    elif method == 'ccn':
        sampler = ClusterCentroids(random_state=42)
    elif method == 'enn':
        sampler = EditedNearestNeighbours()
    elif method == 'renn':
        sampler = RepeatedEditedNearestNeighbours()
    elif method == 'iht':
        sampler = InstanceHardnessThreshold(random_state=42)
    elif method == 'rus':
        sampler = RandomUnderSampler(random_state=42)
    else:
        raise ValueError("Unsupported resampling method")
    return sampler.fit_resample(X, y)

# ------------------------ Feature Selection ------------------------ #
def run_feature_selection(X, y, top_k=30):
    feature_set = set()

    # LASSO
    lasso = LogisticRegressionCV(penalty='l1', solver='saga', cv=5, scoring='roc_auc', max_iter=10000)
    lasso.fit(X, y)
    lasso_selected = X.columns[np.abs(lasso.coef_[0]) > 1e-5]
    feature_set.update(lasso_selected)

    # Random Forest
    rf = RandomForestClassifier(n_estimators=100, random_state=42)
    rf.fit(X, y)
    rf_importances = pd.Series(rf.feature_importances_, index=X.columns)
    rf_selected = rf_importances.nlargest(top_k).index
    feature_set.update(rf_selected)

    # mRMR proxy using mutual_info_classif
    mi_scores = mutual_info_classif(X, y)
    mi_series = pd.Series(mi_scores, index=X.columns)
    mrmr_selected = mi_series.nlargest(top_k).index
    feature_set.update(mrmr_selected)

    return list(feature_set)

# ------------------------ RFE with CV ------------------------ #
def run_rfecv(X, y, features):
    model = LogisticRegression(solver='liblinear')
    rfecv = RFECV(estimator=model, step=1, cv=5, scoring='roc_auc')
    rfecv.fit(X[features], y)
    selected_features = X[features].columns[rfecv.support_]
    return selected_features

# ------------------------ Main Feature Ranking Pipeline ------------------------ #
from tqdm import tqdm
from collections import defaultdict

from sklearn.model_selection import RepeatedStratifiedKFold

from sklearn.preprocessing import StandardScaler

def feature_ranking_pipeline(X, y, repeats=10, k_folds=5, resample_method='smote'):
    rskf = RepeatedStratifiedKFold(n_splits=k_folds, n_repeats=repeats, random_state=42)
    feature_counts = Counter()
    
    # If labels are already integers (0/1), no need to encode
    y_encoded = y.values  # or y.to_numpy()

    total_iterations = repeats * k_folds
    progress_bar = tqdm(total=total_iterations, desc=f"Sampling: {resample_method}", ncols=100)

    for fold_idx, (train_idx, _) in enumerate(rskf.split(X, y_encoded)):
        X_train = X.iloc[train_idx]
        y_train = y_encoded[train_idx]

        # Step 1: Resample
        X_resampled, y_resampled = apply_resampling(X_train, y_train, method=resample_method)
        X_resampled = pd.DataFrame(X_resampled, columns=X.columns)

        # Step 2: Scale (important for LASSO and mutual info)
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X_resampled)
        X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

        # Step 3: Feature Selection (LASSO + RF + mRMR)
        pooled_features = run_feature_selection(X_scaled, y_resampled)

        # Step 4: RFECV
        final_features = run_rfecv(X_scaled, y_resampled, pooled_features)

        # Step 5: Track feature frequency
        feature_counts.update(final_features)

        progress_bar.update(1)

    progress_bar.close()
    return feature_counts



# ------------------------ Loop Over All Sampling Methods ------------------------ #

def run_all_sampling_methods(X, y, repeats=10, k_folds=5):
    sampling_methods = ['smote', 'ccn', 'enn', 'renn', 'iht', 'rus']
    all_feature_rankings = {}

    for method in sampling_methods:
        print(f"\nRunning feature selection with sampling method: {method.upper()}")
        feature_counts = feature_ranking_pipeline(X, y, repeats=repeats, k_folds=k_folds, resample_method=method)
        all_feature_rankings[method] = feature_counts

    return all_feature_rankings

```

```python
def rank_features_from_sampling_results(all_feature_rankings, total_iterations=50):
    """
    Calculate the frequency percentage of each gene per sampling method and compute a total score out of 600.
    Then rank all genes based on this score.

    Parameters:
    - all_feature_rankings: dict {sampling_method: Counter of features}
    - total_iterations: number of iterations per method (default 50)

    Returns:
    - DataFrame sorted by total_score with columns: ['gene', 'smote', 'ccn', 'enn', 'renn', 'iht', 'rus', 'total_score']
    """
    all_genes = set()
    for counter in all_feature_rankings.values():
        all_genes.update(counter.keys())

    data = []
    for gene in all_genes:
        row = {'gene': gene}
        total_score = 0
        for method in all_feature_rankings:
            count = all_feature_rankings[method].get(gene, 0)
            percent = (count / total_iterations) * 100
            row[method] = percent
            total_score += percent
        row['total_score'] = total_score
        data.append(row)

    result_df = pd.DataFrame(data)
    result_df = result_df.sort_values(by='total_score', ascending=False).reset_index(drop=True)
    return result_df

# Example usage:
ranked_genes_df = rank_features_from_sampling_results(all_sampling_feature_rankings)
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def plot_feature_selection_heatmap(all_feature_rankings, top_genes_path, top_n=30):
    """
    Plot a heatmap showing selection frequency of top N genes across resampling methods.

    Args:
        all_feature_rankings (dict): {sampling_method: Counter of features}
        top_genes_path (str): Path to TXT file containing top N genes (one per line)
        top_n (int): Number of top genes to include in the heatmap
    """
    # Load top genes (no header, one gene per line)
    ranked_features = pd.read_csv(top_genes_path, header=None)[0].tolist()

    # Ensure only top_n are used
    feature_list = ranked_features[:top_n]

    # Build heatmap matrix
    method_names = list(all_feature_rankings.keys())
    heatmap_matrix = []
    for gene in feature_list:
        row = []
        for method in method_names:
            count = all_feature_rankings[method].get(gene, 0)
            row.append(count)
        heatmap_matrix.append(row)

    df_heatmap = pd.DataFrame(heatmap_matrix, index=feature_list, columns=method_names)

    # Plot
    plt.figure(figsize=(15, max(8, len(feature_list) // 2)))
    sns.heatmap(df_heatmap, cmap="YlGnBu", annot=True, fmt="d", cbar_kws={'label': 'Selection Frequency'})
    plt.title("Feature Selection Consistency Across Resampling Methods (Top 30 TNBC Genes)")
    plt.xlabel("Resampling Methods")
    plt.ylabel("Top 30 Genes (TNBC Samples Only)")
    plt.tight_layout()
    plt.savefig("feature_selection_heatmap_TNBC_top30.pdf", format='pdf', bbox_inches='tight')
    plt.show()

# Example usage
plot_feature_selection_heatmap(
    all_feature_rankings=all_sampling_feature_rankings,
    top_genes_path="final_ranked_features_top_30_genes_tnbc_samples_only.txt",
    top_n=30
)
```
